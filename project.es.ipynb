#### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

Pista: Puedes ayudarte revisando la demostración multihist de Matplotlib.

```python
# TODO
import matplotlib.pyplot as plt

ds_sur = ds_sur.copy()

ds_sur['price_norm'] = ds_sur.groupby('level5')['price'].transform(
    lambda x: (x - x.min()) / (x.max() - x.min()))

colores = {
    'Fuenlabrada': 'skyblue',
    'Leganés': 'salmon',
    'Getafe': 'mediumseagreen',
    'Alcorcón': 'orchid'}

plt.figure(figsize=(10, 6))
for poblacion in ds_sur['level5'].unique():
    precios = ds_sur[ds_sur['level5'] == poblacion]['price_norm']
    plt.hist(precios, bins=20, alpha=0.6, label=poblacion, color=colores[poblacion])

plt.title("Distribución normalizada de precios por población")
plt.xlabel("Precio normalizado")
plt.ylabel("Frecuencia")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Conclusión

Los histogramas normalizados sugiere que en Getafe y Alcorcón hay mayor diversidad de propiedades en cuanto a precio, posiblemente por diferencias en superficie, ubicación o tipo de vivienda.

#### Ejercicio 19. ¿Qué puedes decir sobre el precio por metro cuadrado (precio/m2) entre los municipios de 'Getafe' y 'Alcorcón'? Debes usar el subconjunto obtenido en la pregunta 14 (★★☆)

Pista: Crea una nueva columna llamada `pps` (price per square en inglés) y luego analiza los valores

```python
# TODO
ds_sur = ds_sur.copy()
ds_sur['pps'] = ds_sur['price'] / ds_sur['surface']

comparacion = ds_sur[ds_sur['level5'].isin(['Getafe', 'Alcorcón'])]

pps_stats = comparacion.groupby('level5')['pps'].agg(['mean', 'median', 'std']).round(2)
print(pps_stats)
```


### Conclusión

Alcorcón presenta un precio por metro cuadrado más alto que Getafe, tanto en media como en mediana.  
Sin embargo, Getafe muestra una mayor dispersión en los precios, lo que indica una oferta más diversa en cuanto a tipo de vivienda, superficie o ubicación.


#### Ejercicio 20. Realiza el mismo gráfico para 4 poblaciones diferentes (columna level5) y colócalos en el mismo gráfico. Debes usar el subconjunto obtenido en la pregunta 14 (★★☆) 
Pista: Haz un diagrama de dispersión de cada población usando subgráficos (subplots).

```python
# TODO
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle("Relación entre superficie y precio por población", fontsize=14)

poblaciones = ['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón']
colores = ['skyblue', 'salmon', 'mediumseagreen', 'orchid']


for i, poblacion in enumerate(poblaciones):
    fila = i // 2
    columna = i % 2
    datos = ds_sur[ds_sur['level5'] == poblacion]
    axs[fila, columna].scatter(datos['surface'], datos['price'], color=colores[i], alpha=0.6)
    axs[fila, columna].set_title(poblacion)
    axs[fila, columna].set_xlabel("Superficie (m²)")
    axs[fila, columna].set_ylabel("Precio (€)")
    axs[fila, columna].grid(True)

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

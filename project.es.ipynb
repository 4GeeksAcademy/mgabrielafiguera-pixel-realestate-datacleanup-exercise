#### Ejercicio 01. ¿Cuál es la casa más cara en todo el dataset? (★☆☆)

`La casa con dirección en Calle del Prado, Nº20 es la más cara y su precio es de 5000000 USD`

#### Ejercicio 03. ¿Cuál es la casa más grande y la más pequeña del dataset? (★☆☆)

Imprime la dirección y el área de las casas seleccionadas. Por ejemplo:

`La casa más grande está ubicada en Calle Gran Vía, Nº38 y su superficie es de 5000 metros`

`La casa más pequeña está ubicada en Calle Mayor, Nº12 y su superficie es de 200 metros`

```python
# TODO
import pandas as pd

ds_filtrado = ds.dropna(subset=['surface'])

mas_grande = ds_filtrado.sort_values(by='surface', ascending=False).iloc[0]

mas_pequena = ds_filtrado.sort_values(by='surface', ascending=True).iloc[0]

print(f"La casa más grande está ubicada en {mas_grande['url_inmueble']} y su superficie es de {int(mas_grande['surface'])} metros")
print(f"La casa más pequeña está ubicada en {mas_pequena['url_inmueble']} y su superficie es de {int(mas_pequena['surface'])} metros")


```

#### Ejercicio 08. Trazar el histograma de los precios para la población (level5 column) de "Arroyomolinos (Madrid)" y explica qué observas (★★☆)

Imprime el histograma de los precios y escribe en la celda del Markdown un breve análisis del trazado.


```python
# TODO: Code
import matplotlib.pyplot as plt

arroyomolinos = ds[ds['level5'] == "Arroyomolinos (Madrid)"]

arroyomolinos['price'].plot.hist(
    bins=20,
    color='skyblue',
    edgecolor='black',
    title='Distribución de precios en Arroyomolinos (Madrid)')

plt.xlabel("Precio en USD")
plt.ylabel("Cantidad de viviendas")
plt.title("Histograma de precios en Arroyomolinos (Madrid)")
plt.grid(True)
plt.show()
```

### Análisis del histograma de precios en Arroyomolinos (Madrid)

El histograma muestra que el precios medio de las viviendas esta entre (200.000–300.000 USD)
Hay pocas casas muy baratas o muy caras.
